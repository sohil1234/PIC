```python
from awg import *
from awg.core import *
import matplotlib.pyplot as plt
import numpy as np
```

```python

model = AWG(
    clad=SiO2, core=Si, subs=SiO2, lambda_c=0.900,  # Central wavelength shifted to 900 nm
    Ni=1, No=4, w=0.350, h=0.250,                  # Reduced waveguide width and height
    N=30, m=100, R=80,                             # Adjusted waveguide count and radius
    d=1.8, g=0.3, do=1.2,                          # Tighter spacing and reduced gap
    wi=1.0, wo=1.0, L0=15                          # Adjusted input/output waveguides and length
)

options = SimulationOptions()
options.PhaseErrorVariance = 0.12
options.ModeType = "solve"
options.PropagationLoss = 1
options.InsertionLoss = 0.5
```

```python
results = Spectrum(model, 0.9, 0.025, Options = options, Samples = 100)
```

```python
plt.plot(results.wavelength, 10*np.log10(results.transmission))
plt.ylim(-40,0)
plt.xlabel('Wavelength ($\mu$m)')
plt.ylabel("Average Transmission (dB)")
plt.show()
```

```python
import numpy as np

# Generate a slightly varied L0 for each waveguide to disrupt periodicity
L0_base = 35
N=25
L0_variation = np.linspace(L0_base - 2, L0_base + 2, N)

# Define the model with a range of L0 values
model_params = {
    'clad': SiO2,
    'core': Si,
    'subs': SiO2,
    'lambda_c': 0.9,
    'Ni': 1,
    'No': 15,
    'w': 0.45,
    'h': 0.25,
    'N': N,
    'm': 100,
    'R': 120,
    'd': 2.2,       # Increased spacing
    'g': 0.5,       # Increased gap to reduce coupling
    'do': 1.5,
    'wi': 3.0,
    'wo': 3.2,
}

# Run the AWG simulation with each L0 variation
transmissions = []
for L0 in L0_variation:
    model = AWG(**model_params, L0=float(L0))
    options = SimulationOptions()
    options.PhaseErrorVariance = 0.02
    options.ModeType = "solve"
    options.PropagationLoss = 0.3
    options.InsertionLoss = 0.3

    # Lower sampling density for faster testing
    results = Spectrum(model, model_params['lambda_c'], 0.05, Options=options, Samples=100)
    transmissions.append(results.transmission)

# Average transmission response across L0 variations
average_transmission = np.mean(transmissions, axis=0)

# Plot results
plt.plot(results.wavelength, 10 * np.log10(average_transmission))
plt.xlabel('Wavelength (µm)')
plt.ylabel('Average Transmission (dB)')
plt.show()
```
